```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
spotify_df.head()
```

```python
# group by genre and calculate the average total streams
avg_streams = spotify_df.groupby("Genre")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_streams
```

Bar Chart

```python
plt.bar(avg_streams.index, avg_streams.values)
plt.show()
```

```python
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel("Genre")
plt.ylabel("Averge Total Streams (Millions)")
plt.title("Average Total Streams by Genre")

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel("Genre")
plt.ylabel("Averge Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

ax = plt.gca()      # get current axes
ax.spines['top'].set_visible(False)    #spines = borders
ax.spines['right'].set_visible(False)

plt.xlabel("Genre")
plt.ylabel("Averge Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
avg_artist_streams = spotify_df.groupby("Artist")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_artist_streams
```

```python
plt.figure(figsize=(10,6))
colors = ["#7b92c9"]*len(avg_artist_streams)

max_index = avg_artist_streams.idxmax()       # gives index of max value
highlight_position = avg_artist_streams.index.get_loc(max_index)
colors[highlight_position] = "#f030ad"

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width = 0.8)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Averge Total Streams (Millions)")
plt.title("Average Total Streams by Artist")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))

special_artist = "Ariana Grande"
colors = ["#f030ad" if artist == special_artist else "gray" for artist in avg_artist_streams.index]


plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width = 0.9)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Averge Total Streams (Millions)")
plt.title("Ariana Grande Ranks 2nd in Total Average Streams")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"], 
    spotify_df["Total Streams (Millions)"],
    color="#08cc15", 
    alpha = 0.8,                  # how solid/transparent the dots are (1 is fully solid)
    s = 25)                       # changes the point size

plt.xlabel("Streams this Month")
plt.ylabel("Total Streams")
plt.title("No Apparent Correlation Between Monthly Listens and Total Streams")

ax = plt.gca()            #get current axes
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

# make the other dots gray
featured_genre = "Pop"
colors = ["#08cc15" if genre == featured_genre else "#d1d5d1" for genre in spotify_df["Genre"]]

plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"], 
    spotify_df["Total Streams (Millions)"],
    color= colors, 
    alpha = 0.8,
    s = 25)

plt.xlabel("Streams this Month")
plt.ylabel("Total Streams")
#plt.title("No Apparent Correlation Between Monthly Listens and Total Streams")

ax = plt.gca()            #get current axes
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# change the font size of the chart title
# a good title should be an observation or a question
# //this title is nonsensical to the data fyi - just an example for practice as is the data in general
ax.set_title("Do Pop Songs Achieve Higher Monthly Listens or Total Streams?", fontsize = 18, pad = 35)

# so we don't have to write this out four times
for spine in ax.spines.values():
    spine.set_linewidth(0.25)
    spine.set_alpha(0.5)

plt.tight_layout()
plt.show()
```

Line Charts

```python
# round - specify the number of decimals
yearly_avg = spotify_df.groupby("Release Year")["Total Streams (Millions)"].mean().round(2)
yearly_avg
```

```python
plt.figure(figsize=(12,6))
plt.plot(
    yearly_avg.index, 
    yearly_avg.values,
    marker = "+",               # //basically the data point markers on the line chart
    linestyle = "--",
    color = "#20a32b",
    linewidth = 0.5)           


plt.xlabel("Release Year")
plt.ylabel("Average Total Streams")
plt.title("Did Listeners Stream Less in 2022?", fontsize = 18, pad = 20)


ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)


ax.axvline(
    2022,
    color = "#063b0a",
    linewidth = 0.5,
    alpha = 0.7)

# adds text to the axes object
# // reminder that we can set color to a variable name in the beginning if using a specific color scheme (branding, etc.)
ax.text(
    2022 + 0.05,
    2225,
    "Dip In Total Streams",
    color = "#063b0a",
    fontsize = 8
)

# want to start the y axis to start at zero and max at 3500
plt.ylim(0, 3500)

plt.tight_layout()
plt.savefig("../plots/2022Streams.png")
plt.show()
```

Pie Charts

```python
category = "Platform Type"
values = "Total Streams (Millions)"

platform_totals = spotify_df.groupby(category)[values].sum()
platform_totals
```

```python
plt.figure(figsize = (8, 8))
plt.pie(
    platform_totals,
    labels = None,
    startangle = 110,                   # pie chart angle
    colors = ['gray', '#1aaa10'],
    autopct = "%1.1f%%"                 # format for the percent value of the wedges - shows 1 decimal place after the number
    )

plt.title("Near Even Split Between Premium and Free Users")


plt.legend(
    labels = platform_totals.index,
    loc = "lower right",
    title = "Platform"
)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8, 8))


wedges, texts, autotexts = ax.pie(
    platform_totals,
    autopct = "%1.1f%%",
    startangle = 110,                   
    colors = ['gray', '#1aaa10'],               
    )

ax.set_title(
    "Near Even Split Between Premium and Free Users",
    fontsize = 30,
    pad = 15,
    loc = "center"
)

for autotext in autotexts:
    autotext.set_fontsize(30)
    autotext.set_color("white")
    autotext.set_weight("bold")


ax.legend(
    wedges,
    platform_totals.index,
    loc = "lower center",
    bbox_to_anchor = (0.5, 0.93),         # allows to move position relative to the chart
    ncol = 2                              # makes legend side by side
)

plt.tight_layout()
plt.show()
```

Histograms

```python
plt.figure(figsize=(10,8))

plt.hist(
    spotify_df["Avg Stream Duration (Min)"],
    bins = 20,
    color = "#2fba28",
    edgecolor = "white"         # gives the bins some separation
)

plt.xlabel("Average Stream Duration (Minutes)")
plt.ylabel("Number of Tracks")
plt.title("Distribution of Average Stream Duration", fontsize = 16, pad = 20)

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(0.5)
    spine.set_alpha(0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig("../plots/AverageStreamHistogram.png")
plt.show()

# //each bar is a "bin"
```

Heatmaps

```python
# correlation is a number between -1 and 1 - tells us the strength and direction of a linear realtionship
# only can use numerical variables of the df
numeric_spotify_df = spotify_df.select_dtypes(include = [np.number])  # np.number gives int and floats (from numpy)
corr = numeric_spotify_df.corr()

im = plt.imshow(corr, vmin = -1, vmax = 1, cmap = "coolwarm")
plt.colorbar(im, label = "Pearson Corellation")

plt.xticks(range(len(corr.columns)), corr.columns, rotation = 45, ha = "right")  # ha = horizontal alignment
plt.yticks(range(len(corr.columns)), corr.columns)

plt.show()
```

Count Plots

```python
plt.figure(figsize=(10, 6))
genre_counts = spotify_df["Genre"].value_counts().sort_values(ascending=False)
genre_counts

plt.bar(genre_counts.index, genre_counts.values)
plt.xticks(rotation = 45, ha = "right")

plt.xlabel("Genre")
plt.ylabel("Count")
plt.title("Number of Tracks by Genre", fontsize = 16, pad = 20)

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(0.5)
    spine.set_alpha(0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```
